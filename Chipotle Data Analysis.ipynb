```python
revenue = 0

for i in range(len(chipotle.item_price)):
    x = int(chipotle.item_price[i])
    revenue += x
    
print(revenue)
```

```python
ordernum = {}

for i in range(len(chipotle.item_name)):
    q = chipotle.item_name[i]
    p = chipotle.quantity[i]
    if q not in ordernum:
        ordernum[q] = p
    else:
        ordernum[q] += p
        
print(ordernum)
print(max(ordernum.values()))
```
